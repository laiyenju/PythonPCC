## 線圖

```
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]

plt.plot(squares)
plt.show()
```

```
squares = [1, 4, 9, 16, 25]

plt.plot(squares, linewidth=3)
plt.title("Square Numbers", fontsize=24)
plt.xlabel("value", fontsize=14)
plt.ylabel("square of value", fontsize=14)

# 刻度標記
plt.tick_params(axis='both', labelsize=14)
plt.show()
```

## 散佈圖 scatter()

```
plt.scatter (2, 4)
plt.show()
```

```
plt.scatter (2, 4, s=100)

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

plt.show()
```

```
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]
plt.scatter(x_values, y_values, s=100)

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

plt.show()
```

```
x_values = list(range(1, 1000))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, s=40)

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

#set the axis's number
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```
x_values = list(range(1, 1000))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, edgecolor='none', s=40)
#edge 是用來消除點的輪廓，讓密集的點連起來不會出現上圖毛邊

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

#set the axis's number
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```
x_values = list(range(1, 1000))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, color='red', edgecolor='none', s=40)

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

#set the axis's number
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

```
x_values = list(range(1, 1000))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, color=(0, 0, 0.8), edgecolor='none', s=40)
#用 RGB 定義顏色

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

#set the axis's number
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

## 色譜（漸層色）colormap

```
#使用色譜 colormap(要注意的是 color 必須以 c 作為變數名稱)
x_values = list(range(1, 1000))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, edgecolor='none', s=40)

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

#set the axis's number
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

## 自動保存圖片

```
x_values = list(range(1, 1000))
y_values = [x**2 for x in x_values]
plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, edgecolor='none', s=40)

plt.title("Scatter Chart", fontsize=22)
plt.xlabel("value", fontsize=12)
plt.ylabel("value squared", fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=10)

plt.axis([0, 1100, 0, 1100000])

#改掉原本的 plit.show()
plt.savefig('squares_plot.png', bbox_inches='tight')
#bbox_inches 表示是否要裁掉空白處，若要保留空白，就不需要另外設定此參數
```

## 題目
#### 15-1 立方：數字的三次方是立方，請繪製一個圖形，顯示前五個整數的立方，再繪製一個圖形，顯示前 5000 個數字的立方

```
import matplotlib.pyplot as plt
x_values=list(range(1,6))
y_values=[x**3 for x in x_values]

plt.title("Cube 1 to 5 Scatter", fontsize=20)
plt.xlabel("number", fontsize=15)
plt.ylabel("cube number", fontsize=15)
plt.tick_params(axis='both', labelsize=8)

plt.scatter(x_values,y_values,s=40)
plt.show()
```

```
import matplotlib.pyplot as plt
x_values=list(range(1,5001))
y_values=[x**3 for x in x_values]

plt.title("Cube 1 to 5000 Scatter", fontsize=20)
plt.xlabel("number", fontsize=15)
plt.ylabel("cube number", fontsize=15)
plt.tick_params(axis='both', labelsize=8)

plt.scatter(x_values,y_values,s=40)
plt.show()
```
